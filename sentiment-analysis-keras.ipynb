```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import re
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics import classification_report, confusion_matrix, roc_auc_score, roc_curve, accuracy_score
from nltk.corpus import stopwords
from nltk.stem import PorterStemmer
import nltk # NLTK verilerini indirmek için eklendi

# NLTK stopwords verisini indirme (eğer daha önce indirilmediyse)
try:
    stopwords.words('english')
except LookupError:
    nltk.download('stopwords')

# Veri setini yükleme
df = pd.read_csv("/kaggle/input/imdb-dataset-of-50k-movie-reviews/IMDB Dataset.csv") 
                                    
df.head()

# Önişleme fonksiyonu
stemmer = PorterStemmer()
stop_words_list = set(stopwords.words('english')) # Değişken adı güncellendi

def preprocess(text):
    text = text.lower()
    text = re.sub(r"[^\w\s]", "", text)  # Noktalama temizle
    tokens = text.split()
    tokens = [stemmer.stem(word) for word in tokens if word not in stop_words_list]
    return " ".join(tokens)

df["cleaned_review"] = df["review"].apply(preprocess)

# Metinleri vektörleştiriyoruz
vectorizer = CountVectorizer(stop_words='english', max_features=5000) # max_features ekleyerek öznitelik sayısını sınırlayabiliriz
x_vec = vectorizer.fit_transform(df["cleaned_review"])
y = df["sentiment"].map({'positive': 1, 'negative': 0}).values # .values ile NumPy array'e çeviriyoruz

print("Öznitelik sayısı:", len(vectorizer.get_feature_names_out()))
print("Vektörlenmiş veri şekli:", x_vec.shape)

# dataset bölme
# CountVectorizer çıktısı sparse matris olduğu için Keras'a vermeden önce yoğun (dense) array'e çeviriyoruz.
x_train, x_test, y_train, y_test = train_test_split(x_vec.toarray(), y, test_size=0.2, random_state=42)

print("Eğitim verisi şekli:", x_train.shape)
print("Test verisi şekli:", x_test.shape)

# Keras ve TensorFlow kütüphanelerini import etme
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout

# Model Oluşturma
# Basit bir tam bağlantılı sinir ağı (Dense Neural Network)
model_keras = Sequential([
    Dense(128, activation='relu', input_shape=(x_train.shape[1],)), # Giriş katmanı, öznitelik sayısı kadar nöron
    Dropout(0.5), # Aşırı öğrenmeyi (overfitting) azaltmak için Dropout katmanı
    Dense(64, activation='relu'),
    Dropout(0.5),
    Dense(1, activation='sigmoid') # Çıkış katmanı, ikili sınıflandırma için sigmoid aktivasyonu
])

# Modeli Derleme
model_keras.compile(optimizer='adam',
                    loss='binary_crossentropy', # İkili sınıflandırma için kayıp fonksiyonu
                    metrics=['accuracy']) # Değerlendirme metriği

model_keras.summary() # Modelin özetini görme

# Modeli Eğitme
# validation_data kullanarak eğitim sırasında test seti üzerindeki performansı izleyebiliriz
history = model_keras.fit(x_train, y_train,
                          epochs=10,       # Eğitim döngüsü sayısı
                          batch_size=512,  # Her bir iterasyonda alınacak örnek sayısı
                          validation_data=(x_test, y_test),
                          verbose=1)       # Eğitim sürecini gösterme (1: göster, 0: gösterme)

# Model Performansını Değerlendirme
loss, accuracy_keras = model_keras.evaluate(x_test, y_test, verbose=0)
print(f"\nKeras Model - Test Kaybı: {loss:.4f}")
print(f"Keras Model - Test Doğruluğu: {accuracy_keras:.4f}")

# Tahminler
y_pred_keras_proba = model_keras.predict(x_test).ravel() # Olasılıklar
y_pred_keras = (y_pred_keras_proba > 0.5).astype(int) # Eşik değere göre sınıflar (0 veya 1)

# Classification Report
print("\nKeras Model - Classification Report:")
print(classification_report(y_test, y_pred_keras, target_names=["negative", "positive"]))

# Confusion Matrix
conf_matrix_keras = confusion_matrix(y_test, y_pred_keras)
plt.figure(figsize=(6, 5))
sns.heatmap(conf_matrix_keras, annot=True, fmt="d", cmap="Blues", xticklabels=["Negative", "Positive"], yticklabels=["Negative", "Positive"])
plt.title("🔍 Keras Model - Confusion Matrix")
plt.xlabel("Predicted Label")
plt.ylabel("True Label")
plt.show()

# ROC Curve ve AUC
fpr_keras, tpr_keras, thresholds_keras = roc_curve(y_test, y_pred_keras_proba)
roc_auc_keras = roc_auc_score(y_test, y_pred_keras_proba)

plt.figure(figsize=(7, 5))
plt.plot(fpr_keras, tpr_keras, color='darkorange', lw=2, label=f'ROC curve (AUC = {roc_auc_keras:.2f})')
plt.plot([0, 1], [0, 1], color='navy', lw=2, linestyle='--')
plt.title('🧠 Keras Model - ROC Curve')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.legend(loc="lower right")
plt.grid(True)
plt.show()

print(f"✅ Keras Model - ROC AUC Score: {roc_auc_keras:.4f}")

# Eğitim geçmişini görselleştirme (isteğe bağlı)
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], label='Eğitim Doğruluğu')
plt.plot(history.history['val_accuracy'], label='Doğrulama Doğruluğu')
plt.title('Keras Model - Doğruluk')
plt.xlabel('Epoch')
plt.ylabel('Doğruluk')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], label='Eğitim Kaybı')
plt.plot(history.history['val_loss'], label='Doğrulama Kaybı')
plt.title('Keras Model - Kayıp')
plt.xlabel('Epoch')
plt.ylabel('Kayıp')
plt.legend()
plt.show()
```
